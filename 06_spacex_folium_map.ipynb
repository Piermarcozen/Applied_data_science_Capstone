```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Carica il dataset
try:
    df = pd.read_csv("dataset_part_1.csv")
    print(f"✅ Dataset caricato: {df.shape}")
except:
    # Se non trovi il file, usa l'URL
    url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/Spacex_Csv.csv"
    df = pd.read_csv(url)
    print(f"✅ Dataset caricato da URL: {df.shape}")

# Mostra le colonne disponibili
print(f"\nColonne disponibili: {df.columns.tolist()}")

# Crea la colonna class da Outcome
bad_outcomes = {'False ASDS', 'False Ocean', 'False RTLS', 'None ASDS', 'None None'}
df['class'] = df['Outcome'].apply(lambda x: 0 if x in bad_outcomes else 1)

print(f"\n✅ Colonna 'class' creata")
print(f"   Successi (class=1): {(df['class'] == 1).sum()}")
print(f"   Fallimenti (class=0): {(df['class'] == 0).sum()}")
```

```python
# Coordinate dei siti di lancio
site_coords = {
    'CCSFS SLC 40': (28.5620, -80.5774),
    'KSC LC 39A': (28.6082, -80.6041),
    'VAFB SLC 4E': (34.6321, -120.6108)
}

# Aggiungi coordinate al DataFrame
df['Lat'] = df['LaunchSite'].map(lambda x: site_coords[x][0])
df['Long'] = df['LaunchSite'].map(lambda x: site_coords[x][1])

print(f"\n✅ Coordinate aggiunte")
print(df[['LaunchSite', 'Lat', 'Long']].drop_duplicates())
```

```python
## Task 1: Mark all launch sites on a map
# Crea mappa centrata sulla Florida
map_center = [28.5, -80.5]
launch_map = folium.Map(location=map_center, zoom_start=5)

# Aggiungi MarkerCluster per gestire molti punti
marker_cluster = MarkerCluster().add_to(launch_map)

# Contatori
success_count = 0
failure_count = 0

# Aggiungi marker per ogni lancio
for idx, row in df.iterrows():
    lat = row['Lat']
    lon = row['Long']
    outcome = row['class']
    
    # Scegli colore e icona
    if outcome == 1:
        icon_color = 'green'
        icon_icon = 'check'
        success_count += 1
    else:
        icon_color = 'red'
        icon_icon = 'times'
        failure_count += 1
    
    # Crea popup
    popup_text = f"""
    <b>Flight: {row['FlightNumber']}</b><br>
    Site: {row['LaunchSite']}<br>
    Date: {row['Date']}<br>
    Orbit: {row['Orbit']}<br>
    Payload: {row['PayloadMass']:.0f} kg<br>
    Outcome: {'✅ Success' if outcome == 1 else '❌ Failure'}
    """
    
    folium.Marker(
        location=[lat, lon],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color=icon_color, icon=icon_icon, prefix='fa'),
        tooltip=f"Flight {row['FlightNumber']}"
    ).add_to(marker_cluster)

# Mostra la mappa
display(launch_map)

print(f"\n✅ Mappa creata con {len(df)} lanci")
print(f"   🟢 Successi: {success_count}")
print(f"   🔴 Fallimenti: {failure_count}")
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`

```python
# TASK 3: Calculate the distances between a launch site to its proximities
"""
TASK 3: Calculate the distances between a launch site to its proximities
Calcolo delle distanze dai siti di lancio a: costa, equatore, città, aeroporti
"""

import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
from math import radians, cos, sin, asin, sqrt

print("="*80)
print("🌍 TASK 3: CALCOLO DISTANZE SITI DI LANCIO")
print("="*80)

# ============================================================================
# 1. FUNZIONE PER CALCOLO DISTANZA (Haversine Formula)
# ============================================================================

def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calcola la distanza in km tra due punti sulla Terra
    usando la formula di Haversine
    """
    # Converte gradi in radianti
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Formula di Haversine
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    
    # Raggio della Terra in km
    r = 6371
    return c * r


# ============================================================================
# 2. DEFINIZIONE DEI SITI DI LANCIO
# ============================================================================

print("\n📍 SITI DI LANCIO")
print("-"*50)

launch_sites = {
    'CCSFS SLC 40': {'lat': 28.5620, 'lon': -80.5774, 'coast': 'Atlantic Ocean'},
    'KSC LC 39A': {'lat': 28.6082, 'lon': -80.6041, 'coast': 'Atlantic Ocean'},
    'VAFB SLC 4E': {'lat': 34.6321, 'lon': -120.6108, 'coast': 'Pacific Ocean'}
}

for site, coords in launch_sites.items():
    print(f"  {site}: ({coords['lat']:.4f}, {coords['lon']:.4f}) - {coords['coast']}")

# ============================================================================
# 3. DEFINIZIONE DEI PUNTI DI RIFERIMENTO
# ============================================================================

print("\n📌 PUNTI DI RIFERIMENTO")
print("-"*50)

# Costa
coast_points = {
    'Atlantic_Ocean': {'lat': 28.5, 'lon': -80.0, 'name': 'Atlantic Ocean (Cape Canaveral)'},
    'Pacific_Ocean': {'lat': 34.7, 'lon': -120.7, 'name': 'Pacific Ocean (Vandenberg)'}
}

# Equatore
equator = {'lat': 0, 'lon': -80.0, 'name': 'Equator'}

# Città principali
cities = {
    'Orlando': {'lat': 28.5383, 'lon': -81.3792},
    'Miami': {'lat': 25.7617, 'lon': -80.1918},
    'Tampa': {'lat': 27.9506, 'lon': -82.4572},
    'Jacksonville': {'lat': 30.3322, 'lon': -81.6557},
    'Cape_Canaveral': {'lat': 28.3883, 'lon': -80.6038},
    'Los_Angeles': {'lat': 34.0522, 'lon': -118.2437},
    'San_Francisco': {'lat': 37.7749, 'lon': -122.4194},
    'Santa_Barbara': {'lat': 34.4208, 'lon': -119.6982}
}

# Aeroporti principali
airports = {
    'MCO': {'lat': 28.4294, 'lon': -81.3089, 'name': 'Orlando International'},
    'MIA': {'lat': 25.7932, 'lon': -80.2906, 'name': 'Miami International'},
    'TPA': {'lat': 27.9755, 'lon': -82.5332, 'name': 'Tampa International'},
    'JAX': {'lat': 30.4941, 'lon': -81.6879, 'name': 'Jacksonville International'},
    'LAX': {'lat': 33.9416, 'lon': -118.4085, 'name': 'Los Angeles International'},
    'SFO': {'lat': 37.6213, 'lon': -122.3790, 'name': 'San Francisco International'},
    'SBA': {'lat': 34.4262, 'lon': -119.8404, 'name': 'Santa Barbara Airport'},
    'VAFB': {'lat': 34.7333, 'lon': -120.5667, 'name': 'Vandenberg AFB Airport'}
}

print("  ✅ Punti di riferimento definiti")

# ============================================================================
# 4. CALCOLO DISTANZE
# ============================================================================

print("\n📏 CALCOLO DISTANZE IN CORSO...")
print("-"*50)

# Lista per memorizzare i risultati
distances_data = []

for site_name, site_coords in launch_sites.items():
    site_lat = site_coords['lat']
    site_lon = site_coords['lon']
    
    print(f"\n  Calcolo per {site_name}...")
    
    # Distanza dall'Equatore
    dist_equator = haversine_distance(site_lat, site_lon, equator['lat'], equator['lon'])
    print(f"    → Distanza Equatore: {dist_equator:.1f} km")
    
    # Distanza dalla costa più vicina
    if site_coords['coast'] == 'Atlantic Ocean':
        coast_point = coast_points['Atlantic_Ocean']
        dist_coast = haversine_distance(site_lat, site_lon, coast_point['lat'], coast_point['lon'])
    else:
        coast_point = coast_points['Pacific_Ocean']
        dist_coast = haversine_distance(site_lat, site_lon, coast_point['lat'], coast_point['lon'])
    print(f"    → Distanza Costa ({site_coords['coast']}): {dist_coast:.1f} km")
    
    # Distanze dalle città
    city_distances = {}
    for city_name, city_coords in cities.items():
        dist = haversine_distance(site_lat, site_lon, city_coords['lat'], city_coords['lon'])
        city_distances[city_name] = dist
    
    # Trova la città più vicina
    nearest_city = min(city_distances, key=city_distances.get)
    nearest_city_dist = city_distances[nearest_city]
    print(f"    → Città più vicina: {nearest_city} ({nearest_city_dist:.1f} km)")
    
    # Distanze dagli aeroporti
    airport_distances = {}
    for airport_code, airport_coords in airports.items():
        dist = haversine_distance(site_lat, site_lon, airport_coords['lat'], airport_coords['lon'])
        airport_distances[airport_code] = dist
    
    # Trova l'aeroporto più vicino
    nearest_airport = min(airport_distances, key=airport_distances.get)
    nearest_airport_dist = airport_distances[nearest_airport]
    nearest_airport_name = airports[nearest_airport]['name']
    print(f"    → Aeroporto più vicino: {nearest_airport_name} ({nearest_airport_dist:.1f} km)")
    
    # Salva i risultati
    distances_data.append({
        'Launch_Site': site_name,
        'Latitude': site_lat,
        'Longitude': site_lon,
        'Coast': site_coords['coast'],
        'Distance_to_Equator_km': round(dist_equator, 1),
        'Distance_to_Coast_km': round(dist_coast, 1),
        'Nearest_City': nearest_city,
        'Distance_to_Nearest_City_km': round(nearest_city_dist, 1),
        'Nearest_Airport': nearest_airport_name,
        'Distance_to_Nearest_Airport_km': round(nearest_airport_dist, 1),
        'Distance_to_Orlando_km': round(city_distances.get('Orlando', 0), 1),
        'Distance_to_Miami_km': round(city_distances.get('Miami', 0), 1),
        'Distance_to_Cape_Canaveral_km': round(city_distances.get('Cape_Canaveral', 0), 1),
        'Distance_to_Los_Angeles_km': round(city_distances.get('Los_Angeles', 0), 1)
    })

# Crea DataFrame
distances_df = pd.DataFrame(distances_data)

# ============================================================================
# 5. VISUALIZZAZIONE TABELLA DISTANZE
# ============================================================================

print("\n" + "="*80)
print("📊 TABELLA RIASSUNTIVA DISTANZE")
print("="*80)

# Selezione colonne principali per visualizzazione
display_cols = ['Launch_Site', 'Coast', 'Distance_to_Equator_km', 'Distance_to_Coast_km', 
                'Nearest_City', 'Distance_to_Nearest_City_km', 'Nearest_Airport', 'Distance_to_Nearest_Airport_km']

print(distances_df[display_cols].to_string(index=False))

# ============================================================================
# 6. STATISTICHE E ANALISI
# ============================================================================

print("\n" + "="*80)
print("📈 ANALISI STATISTICA")
print("="*80)

print(f"\n📍 SITO PIÙ VICINO ALL'EQUATORE:")
min_equator = distances_df.loc[distances_df['Distance_to_Equator_km'].idxmin()]
print(f"   {min_equator['Launch_Site']}: {min_equator['Distance_to_Equator_km']:.1f} km")

print(f"\n📍 SITO PIÙ VICINO ALLA COSTA:")
min_coast = distances_df.loc[distances_df['Distance_to_Coast_km'].idxmin()]
print(f"   {min_coast['Launch_Site']}: {min_coast['Distance_to_Coast_km']:.1f} km")

print(f"\n📍 SITO CON AEROPORTO PIÙ VICINO:")
min_airport = distances_df.loc[distances_df['Distance_to_Nearest_Airport_km'].idxmin()]
print(f"   {min_airport['Launch_Site']}: {min_airport['Nearest_Airport']} ({min_airport['Distance_to_Nearest_Airport_km']:.1f} km)")

print(f"\n📊 DISTANZA MEDIA DALL'EQUATORE: {distances_df['Distance_to_Equator_km'].mean():.1f} km")
print(f"📊 DISTANZA MEDIA DALLA COSTA: {distances_df['Distance_to_Coast_km'].mean():.1f} km")

# ============================================================================
# 7. CREAZIONE MAPPA INTERATTIVA
# ============================================================================

print("\n🗺️ CREAZIONE MAPPA INTERATTIVA...")
print("-"*50)

# Crea mappa centrata sugli USA
map_center = [31.5, -100]
proximity_map = folium.Map(location=map_center, zoom_start=4)

# Aggiungi i siti di lancio (cerchi rossi)
for site_name, site_coords in launch_sites.items():
    folium.CircleMarker(
        location=[site_coords['lat'], site_coords['lon']],
        radius=10,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.7,
        popup=folium.Popup(f"<b>{site_name}</b><br>Lat: {site_coords['lat']}<br>Lon: {site_coords['lon']}", max_width=200),
        tooltip=site_name
    ).add_to(proximity_map)

# Aggiungi le città
for city_name, city_coords in cities.items():
    folium.Marker(
        location=[city_coords['lat'], city_coords['lon']],
        popup=f"<b>{city_name}</b>",
        icon=folium.Icon(color='blue', icon='info-sign', prefix='fa'),
        tooltip=city_name
    ).add_to(proximity_map)

# Aggiungi gli aeroporti
for airport_code, airport_coords in airports.items():
    folium.Marker(
        location=[airport_coords['lat'], airport_coords['lon']],
        popup=f"<b>{airport_coords['name']}</b>",
        icon=folium.Icon(color='orange', icon='plane', prefix='fa'),
        tooltip=airport_coords['name']
    ).add_to(proximity_map)

# Aggiungi l'equatore (linea)
folium.PolyLine(
    locations=[[0, -130], [0, -70]],
    color='blue',
    weight=2,
    opacity=0.5,
    popup='Equator'
).add_to(proximity_map)

# Aggiungi la costa (punti di riferimento)
for coast_name, coast_coords in coast_points.items():
    folium.Marker(
        location=[coast_coords['lat'], coast_coords['lon']],
        popup=f"<b>{coast_coords['name']}</b>",
        icon=folium.Icon(color='lightblue', icon='water', prefix='fa'),
        tooltip=coast_coords['name']
    ).add_to(proximity_map)

# Mostra la mappa
display(proximity_map)

# ============================================================================
# 8. DOMANDE DI ANALISI
# ============================================================================

print("\n" + "="*80)
print("❓ ANALISI E RISPOSTE ALLE DOMANDE")
print("="*80)

print("""
1. Are all launch sites in proximity to the Equator line?
   → NO. I siti in Florida (CCSFS e KSC) sono a circa 28.5°N (3.170 km dall'Equatore),
     mentre VAFB in California è a 34.6°N (3.850 km dall'Equatore).
     I siti in Florida sono più vicini all'Equatore, il che offre un vantaggio
     in termini di velocità di rotazione terrestre per i lanci verso est.

2. Are all launch sites in very close proximity to the coast?
   → SÌ. Tutti i siti sono posizionati sulla costa (distanza < 15 km):
     - CCSFS SLC 40: a ~20 km dalla costa atlantica
     - KSC LC 39A: a ~22 km dalla costa atlantica
     - VAFB SLC 4E: a ~5 km dalla costa pacifica
     
     Questo posizionamento è strategico per:
     - Sicurezza: i razzi cadono in mare
     - Recupero: facilità di recupero dei booster via drone ship
     - Traiettorie: lanci orientati verso l'oceano

3. Quali sono le implicazioni logistiche?
   → I siti sono ben collegati a infrastrutture di trasporto:
     - Florida: vicino a Orlando (aeroporto internazionale a ~60-70 km)
     - California: vicino a Santa Barbara e Vandenberg AFB Airport
     - Ciò facilita il trasporto dei componenti dei razzi
""")

# ============================================================================
# 9. SALVATAGGIO DATI (OPZIONALE)
# ============================================================================

print("\n💾 SALVATAGGIO DATI...")
print("-"*50)

# Salva il DataFrame in CSV
distances_df.to_csv("spacex_launch_site_distances.csv", index=False)
print("✅ File salvato: spacex_launch_site_distances.csv")

# ============================================================================
# 10. RIEPILOGO FINALE
# ============================================================================

print("\n" + "="*80)
print("✅ TASK 3 COMPLETATO CON SUCCESSO!")
print("="*80)
print(f"\n📊 RIEPILOGO:")
print(f"   • Siti analizzati: {len(distances_df)}")
print(f"   • Punti di riferimento: {len(cities)} città, {len(airports)} aeroporti")
print(f"   • Distanza media dall'Equatore: {distances_df['Distance_to_Equator_km'].mean():.1f} km")
print(f"   • Distanza media dalla costa: {distances_df['Distance_to_Coast_km'].mean():.1f} km")
print(f"\n🗺️ Mappa interattiva creata con {len(launch_sites)} siti, {len(cities)} città e {len(airports)} aeroporti")
print("\n🚀 Pronto per i prossimi task!")
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)
"""
TASK 3: PolyLine tra siti di lancio e punti di interesse
Usando MousePosition per trovare coordinate di costa, città, ferrovia, autostrada
"""

import folium
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# ============================================================================
# 1. FUNZIONE PER CALCOLO DISTANZA
# ============================================================================

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calcola la distanza in km tra due punti sulla Terra
    """
    R = 6373.0
    
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c
    return distance


# ============================================================================
# 2. DEFINIZIONE SITI DI LANCIO
# ============================================================================

launch_sites = {
    'CCSFS SLC 40': {
        'lat': 28.5620, 
        'lon': -80.5774,
        'name': 'Cape Canaveral SLC 40',
        'color': 'red'
    },
    'KSC LC 39A': {
        'lat': 28.6082, 
        'lon': -80.6041,
        'name': 'Kennedy Space Center LC 39A',
        'color': 'red'
    },
    'VAFB SLC 4E': {
        'lat': 34.6321, 
        'lon': -120.6108,
        'name': 'Vandenberg SLC 4E',
        'color': 'red'
    }
}

# ============================================================================
# 3. PUNTI DI INTERESSE (MODIFICA CON LE COORDINATE TROVATE CON MOUSEPOSITION)
# ============================================================================

# Per CCSFS SLC 40 (Cape Canaveral) - MODIFICA QUESTE COORDINATE!
ccsfs_points = {
    'Coastline': {
        'lat': 28.56367,    # <-- MODIFICA con MousePosition
        'lon': -80.57163,   # <-- MODIFICA con MousePosition
        'color': 'blue',
        'icon': '🌊',
        'description': 'Atlantic Ocean Coast'
    },
    'City (Cape Canaveral)': {
        'lat': 28.3883,     # <-- MODIFICA con MousePosition
        'lon': -80.6038,    # <-- MODIFICA con MousePosition
        'color': 'green',
        'icon': '🏙️',
        'description': 'Cape Canaveral City'
    },
    'Railway': {
        'lat': 28.5280,     # <-- MODIFICA con MousePosition
        'lon': -80.6620,    # <-- MODIFICA con MousePosition
        'color': 'purple',
        'icon': '🚂',
        'description': 'Florida East Coast Railway'
    },
    'Highway (SR 401)': {
        'lat': 28.5850,     # <-- MODIFICA con MousePosition
        'lon': -80.6200,    # <-- MODIFICA con MousePosition
        'color': 'orange',
        'icon': '🛣️',
        'description': 'Highway SR 401'
    },
    'Port Canaveral': {
        'lat': 28.4150,     # <-- MODIFICA con MousePosition
        'lon': -80.6050,    # <-- MODIFICA con MousePosition
        'color': 'darkblue',
        'icon': '⚓',
        'description': 'Port Canaveral'
    }
}

# Per KSC LC 39A (Kennedy Space Center) - MODIFICA QUESTE COORDINATE!
ksc_points = {
    'Coastline': {
        'lat': 28.60900,    # <-- MODIFICA con MousePosition
        'lon': -80.56500,   # <-- MODIFICA con MousePosition
        'color': 'blue',
        'icon': '🌊',
        'description': 'Atlantic Ocean Coast'
    },
    'City (Titusville)': {
        'lat': 28.6122,     # <-- MODIFICA con MousePosition
        'lon': -80.8076,    # <-- MODIFICA con MousePosition
        'color': 'green',
        'icon': '🏙️',
        'description': 'Titusville'
    },
    'Highway (NASA Causeway)': {
        'lat': 28.5300,     # <-- MODIFICA con MousePosition
        'lon': -80.6200,    # <-- MODIFICA con MousePosition
        'color': 'orange',
        'icon': '🛣️',
        'description': 'NASA Causeway'
    },
    'Vehicle Assembly Building': {
        'lat': 28.5860,     # <-- MODIFICA con MousePosition
        'lon': -80.6510,    # <-- MODIFICA con MousePosition
        'color': 'darkred',
        'icon': '🏭',
        'description': 'VAB'
    }
}

# Per VAFB SLC 4E (Vandenberg) - MODIFICA QUESTE COORDINATE!
vafb_points = {
    'Coastline': {
        'lat': 34.63250,    # <-- MODIFICA con MousePosition
        'lon': -120.63500,  # <-- MODIFICA con MousePosition
        'color': 'blue',
        'icon': '🌊',
        'description': 'Pacific Ocean Coast'
    },
    'City (Lompoc)': {
        'lat': 34.6392,     # <-- MODIFICA con MousePosition
        'lon': -120.4579,   # <-- MODIFICA con MousePosition
        'color': 'green',
        'icon': '🏙️',
        'description': 'Lompoc'
    },
    'Railway': {
        'lat': 34.6800,     # <-- MODIFICA con MousePosition
        'lon': -120.5800,   # <-- MODIFICA con MousePosition
        'color': 'purple',
        'icon': '🚂',
        'description': 'Union Pacific Railway'
    },
    'Highway (CA-1)': {
        'lat': 34.6500,     # <-- MODIFICA con MousePosition
        'lon': -120.6000,   # <-- MODIFICA con MousePosition
        'color': 'orange',
        'icon': '🛣️',
        'description': 'Highway CA-1'
    }
}

# Raggruppa tutti i punti per sito
all_points = {
    'CCSFS SLC 40': ccsfs_points,
    'KSC LC 39A': ksc_points,
    'VAFB SLC 4E': vafb_points
}

# ============================================================================
# 4. CREAZIONE MAPPA CON MOUSE POSITION
# ============================================================================

# Crea la mappa centrata sugli USA
site_map = folium.Map(location=[31.5, -100], zoom_start=5)

# Formattatore per le coordinate
formatter = "function(num) {return L.Util.formatNum(num, 5);};"

# Aggiungi MousePosition
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# ============================================================================
# 5. AGGIUNTA MARKER PER I SITI DI LANCIO
# ============================================================================

for site_key, site_info in launch_sites.items():
    folium.Marker(
        location=[site_info['lat'], site_info['lon']],
        popup=f"""
        <b>🚀 {site_info['name']}</b><br>
        Latitude: {site_info['lat']:.5f}<br>
        Longitude: {site_info['lon']:.5f}
        """,
        icon=folium.Icon(color=site_info['color'], icon='rocket', prefix='fa'),
        tooltip=site_info['name']
    ).add_to(site_map)

# ============================================================================
# 6. AGGIUNTA MARKER PER PUNTI DI INTERESSE E POLYLINE
# ============================================================================

print("\n" + "="*80)
print("📍 CREAZIONE MARKER E POLYLINE PER PUNTI DI INTERESSE")
print("="*80)

for site_key, points in all_points.items():
    site_lat = launch_sites[site_key]['lat']
    site_lon = launch_sites[site_key]['lon']
    site_name = launch_sites[site_key]['name']
    
    print(f"\n📍 {site_name}:")
    
    for point_name, point_info in points.items():
        point_lat = point_info['lat']
        point_lon = point_info['lon']
        point_color = point_info['color']
        point_icon = point_info['icon']
        point_desc = point_info['description']
        
        # Calcola la distanza
        distance = calculate_distance(site_lat, site_lon, point_lat, point_lon)
        print(f"   → {point_icon} {point_name}: {distance:.2f} km")
        
        # ====================================================================
        # 6a. AGGIUNGI MARKER PER IL PUNTO DI INTERESSE
        # ====================================================================
        
        # Marker con DivIcon per visualizzare la distanza
        distance_marker = folium.Marker(
            location=[point_lat, point_lon],
            icon=DivIcon(
                icon_size=(150, 36),
                icon_anchor=(75, 18),
                html=f'''
                <div style="
                    font-size: 11px; 
                    font-weight: bold;
                    color: {point_color}; 
                    background-color: white;
                    border: 2px solid {point_color};
                    border-radius: 5px;
                    padding: 2px 6px;
                    white-space: nowrap;
                    box-shadow: 2px 2px 4px rgba(0,0,0,0.2);
                ">
                    {point_icon} {distance:.2f} KM
                </div>
                ''',
            ),
            tooltip=f"{point_name}: {distance:.2f} km"
        )
        distance_marker.add_to(site_map)
        
        # ====================================================================
        # 6b. AGGIUNGI POLYLINE TRA SITO E PUNTO DI INTERESSE
        # ====================================================================
        
        # Crea la PolyLine
        polyline = folium.PolyLine(
            locations=[[site_lat, site_lon], [point_lat, point_lon]],
            color=point_color,
            weight=2,
            opacity=0.7,
            popup=f"""
            <b>{site_name} → {point_name}</b><br>
            Distanza: {distance:.2f} km<br>
            {point_desc}<br>
            Coordinate: ({point_lat:.5f}, {point_lon:.5f})
            """,
            tooltip=f"{point_name}: {distance:.2f} km"
        )
        polyline.add_to(site_map)
        
        # ====================================================================
        # 6c. AGGIUNGI CERCHIO SUL PUNTO DI INTERESSE (OPZIONALE)
        # ====================================================================
        
        folium.Circle(
            location=[point_lat, point_lon],
            radius=500,
            color=point_color,
            fill=True,
            fill_color=point_color,
            fill_opacity=0.3,
            popup=f"{point_icon} {point_name}<br>Distanza: {distance:.2f} km",
            tooltip=point_name
        ).add_to(site_map)

# ============================================================================
# 7. AGGIUNGI LEGENDA
# ============================================================================

legend_html = '''
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: 320px; height: auto;
            border:2px solid grey; z-index:9999; font-size:12px;
            background-color:white; opacity:0.95; padding: 10px;
            border-radius: 8px;
            box-shadow: 3px 3px 5px rgba(0,0,0,0.2);">
<b>📍 LEGENDA</b><br>
🔴 <span style="color:red">Sito di lancio</span><br>
🌊 <span style="color:blue">Linea/cerchio blu = Costa</span><br>
🏙️ <span style="color:green">Linea/cerchio verde = Città</span><br>
🚂 <span style="color:purple">Linea/cerchio viola = Ferrovia</span><br>
🛣️ <span style="color:orange">Linea/cerchio arancione = Autostrada</span><br>
⚓ <span style="color:darkblue">Linea/cerchio blu scuro = Porto</span><br>
🏭 <span style="color:darkred">Linea/cerchio rosso scuro = Strutture</span><br>
🖱️ <b>MousePosition</b> in alto a destra<br>
<br>
<b>📝 ISTRUZIONI:</b><br>
1. Usa ZOOM per avvicinarti ai siti<br>
2. Muovi il mouse per vedere coordinate<br>
3. MODIFICA le coordinate nei dizionari<br>
   `ccsfs_points`, `ksc_points`, `vafb_points`<br>
4. Riesegui la cella per aggiornare
</div>
'''

site_map.get_root().html.add_child(folium.Element(legend_html))

# ============================================================================
# 8. MOSTRA MAPPA
# ============================================================================

print("\n" + "="*80)
print("🗺️ MAPPA CREATA CON SUCCESSO!")
print("="*80)
print("\n✅ Elementi aggiunti alla mappa:")
print("   • 🔴 Marker per i siti di lancio")
print("   • 🌊 Linee e marker per la costa")
print("   • 🏙️ Linee e marker per le città")
print("   • 🚂 Linee e marker per le ferrovie")
print("   • 🛣️ Linee e marker per le autostrade")
print("   • ⚓ Linee e marker per i porti")
print("   • 🏭 Linee e marker per strutture")
print("   • 🖱️ MousePosition in alto a destra")
print("   • 📋 Legenda in basso a sinistra")

# Mostra la mappa
display(site_map)

# ============================================================================
# 9. RIEPILOGO DISTANZE
# ============================================================================

print("\n" + "="*80)
print("📊 RIEPILOGO DISTANZE PER SITO")
print("="*80)

for site_key, points in all_points.items():
    site_lat = launch_sites[site_key]['lat']
    site_lon = launch_sites[site_key]['lon']
    site_name = launch_sites[site_key]['name']
    
    print(f"\n📍 {site_name}:")
    
    for point_name, point_info in points.items():
        point_lat = point_info['lat']
        point_lon = point_info['lon']
        point_icon = point_info['icon']
        
        distance = calculate_distance(site_lat, site_lon, point_lat, point_lon)
        print(f"   → {point_icon} {point_name}: {distance:.2f} km")

print("\n" + "="*80)
print("✅ TASK COMPLETATO!")
print("="*80)
print("\n💡 SUGGERIMENTI PER PERSONALIZZARE:")
print("   • Usa il MousePosition per trovare coordinate precise")
print("   • Modifica i dizionari `ccsfs_points`, `ksc_points`, `vafb_points`")
print("   • Aggiungi nuovi punti di interesse (aeroporti, stazioni, ecc.)")
print("   • Cambia i colori delle linee modificando il parametro 'color'")
print("   • Riesegui la cella per aggiornare marker, distanze e PolyLine")
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site


import folium
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# ============================================================================
# 1. FUNZIONE PER CALCOLO DISTANZA
# ============================================================================

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calcola la distanza in km tra due punti sulla Terra
    """
    R = 6373.0
    
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c
    return distance


# ============================================================================
# 2. DEFINIZIONE SITI DI LANCIO
# ============================================================================

launch_sites = {
    'CCSFS SLC 40': {
        'lat': 28.5620, 
        'lon': -80.5774,
        'name': '🚀 Cape Canaveral SLC 40',
        'color': 'red'
    },
    'KSC LC 39A': {
        'lat': 28.6082, 
        'lon': -80.6041,
        'name': '🚀 Kennedy Space Center LC 39A',
        'color': 'red'
    },
    'VAFB SLC 4E': {
        'lat': 34.6321, 
        'lon': -120.6108,
        'name': '🚀 Vandenberg SLC 4E',
        'color': 'red'
    }
}

# ============================================================================
# 3. PUNTI DI INTERESSE PER CCSFS SLC 40 (MODIFICA CON MOUSEPOSITION)
# ============================================================================

ccsfs_poi = {
    '🌊 Coastline (Atlantic Ocean)': {
        'lat': 28.56367, 'lon': -80.57163,
        'color': 'blue', 'icon': '🌊', 'line_color': 'lightblue'
    },
    '🏙️ City (Cape Canaveral)': {
        'lat': 28.3883, 'lon': -80.6038,
        'color': 'green', 'icon': '🏙️', 'line_color': 'lightgreen'
    },
    '🚂 Railway (Florida East Coast)': {
        'lat': 28.5280, 'lon': -80.6620,
        'color': 'purple', 'icon': '🚂', 'line_color': 'violet'
    },
    '🛣️ Highway (SR 401)': {
        'lat': 28.5850, 'lon': -80.6200,
        'color': 'orange', 'icon': '🛣️', 'line_color': 'orange'
    },
    '⚓ Port Canaveral': {
        'lat': 28.4150, 'lon': -80.6050,
        'color': 'darkblue', 'icon': '⚓', 'line_color': 'blue'
    },
    '✈️ Airport (Orlando Intl)': {
        'lat': 28.4294, 'lon': -81.3089,
        'color': 'cadetblue', 'icon': '✈️', 'line_color': 'cadetblue'
    }
}

# ============================================================================
# 4. PUNTI DI INTERESSE PER KSC LC 39A (MODIFICA CON MOUSEPOSITION)
# ============================================================================

ksc_poi = {
    '🌊 Coastline (Atlantic Ocean)': {
        'lat': 28.60900, 'lon': -80.56500,
        'color': 'blue', 'icon': '🌊', 'line_color': 'lightblue'
    },
    '🏙️ City (Titusville)': {
        'lat': 28.6122, 'lon': -80.8076,
        'color': 'green', 'icon': '🏙️', 'line_color': 'lightgreen'
    },
    '🛣️ Highway (NASA Causeway)': {
        'lat': 28.5300, 'lon': -80.6200,
        'color': 'orange', 'icon': '🛣️', 'line_color': 'orange'
    },
    '🏭 Vehicle Assembly Building': {
        'lat': 28.5860, 'lon': -80.6510,
        'color': 'darkred', 'icon': '🏭', 'line_color': 'red'
    },
    '✈️ Airport (Space Coast)': {
        'lat': 28.4761, 'lon': -80.7447,
        'color': 'cadetblue', 'icon': '✈️', 'line_color': 'cadetblue'
    }
}

# ============================================================================
# 5. PUNTI DI INTERESSE PER VAFB SLC 4E (MODIFICA CON MOUSEPOSITION)
# ============================================================================

vafb_poi = {
    '🌊 Coastline (Pacific Ocean)': {
        'lat': 34.63250, 'lon': -120.63500,
        'color': 'blue', 'icon': '🌊', 'line_color': 'lightblue'
    },
    '🏙️ City (Lompoc)': {
        'lat': 34.6392, 'lon': -120.4579,
        'color': 'green', 'icon': '🏙️', 'line_color': 'lightgreen'
    },
    '🚂 Railway (Union Pacific)': {
        'lat': 34.6800, 'lon': -120.5800,
        'color': 'purple', 'icon': '🚂', 'line_color': 'violet'
    },
    '🛣️ Highway (CA-1)': {
        'lat': 34.6500, 'lon': -120.6000,
        'color': 'orange', 'icon': '🛣️', 'line_color': 'orange'
    },
    '✈️ Airport (Santa Barbara)': {
        'lat': 34.4262, 'lon': -119.8404,
        'color': 'cadetblue', 'icon': '✈️', 'line_color': 'cadetblue'
    }
}

# Raggruppa tutti i punti per sito
all_poi = {
    'CCSFS SLC 40': ccsfs_poi,
    'KSC LC 39A': ksc_poi,
    'VAFB SLC 4E': vafb_poi
}

# ============================================================================
# 6. CREAZIONE MAPPA CON MOUSE POSITION
# ============================================================================

# Crea la mappa
site_map = folium.Map(location=[31.5, -100], zoom_start=5)

# Formattatore per le coordinate
formatter = "function(num) {return L.Util.formatNum(num, 5);};"

# Aggiungi MousePosition
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# ============================================================================
# 7. AGGIUNTA MARKER PER I SITI DI LANCIO
# ============================================================================

for site_key, site_info in launch_sites.items():
    folium.Marker(
        location=[site_info['lat'], site_info['lon']],
        popup=f"""
        <b>{site_info['name']}</b><br>
        📍 Latitude: {site_info['lat']:.5f}<br>
        📍 Longitude: {site_info['lon']:.5f}
        """,
        icon=folium.Icon(color=site_info['color'], icon='rocket', prefix='fa'),
        tooltip=site_info['name']
    ).add_to(site_map)

# ============================================================================
# 8. AGGIUNTA MARKER E LINEE PER PUNTI DI INTERESSE
# ============================================================================

print("\n" + "="*80)
print("📍 CREAZIONE MARKER CON DISTANZA E LINEE")
print("="*80)

for site_key, poi_dict in all_poi.items():
    site_lat = launch_sites[site_key]['lat']
    site_lon = launch_sites[site_key]['lon']
    site_name = launch_sites[site_key]['name']
    
    print(f"\n📍 {site_name}:")
    
    for poi_name, poi_info in poi_dict.items():
        poi_lat = poi_info['lat']
        poi_lon = poi_info['lon']
        poi_color = poi_info['color']
        poi_icon = poi_info['icon']
        line_color = poi_info['line_color']
        
        # Calcola la distanza
        distance = calculate_distance(site_lat, site_lon, poi_lat, poi_lon)
        print(f"   → {poi_icon} {poi_name}: {distance:.2f} km")
        
        # ====================================================================
        # 8a. CREA MARKER CON DISTANZA (DIVICON)
        # ====================================================================
        
        marker_html = f'''
        <div style="
            font-size: 11px; 
            font-weight: bold;
            color: {poi_color}; 
            background-color: white;
            border: 2px solid {poi_color};
            border-radius: 5px;
            padding: 2px 6px;
            white-space: nowrap;
            box-shadow: 2px 2px 4px rgba(0,0,0,0.2);
            font-family: monospace;
        ">
            {poi_icon} {distance:.2f} KM
        </div>
        '''
        
        distance_marker = folium.Marker(
            location=[poi_lat, poi_lon],
            icon=DivIcon(
                icon_size=(150, 36),
                icon_anchor=(75, 18),
                html=marker_html
            ),
            tooltip=f"{poi_name}: {distance:.2f} km dal sito di lancio"
        )
        distance_marker.add_to(site_map)
        
        # ====================================================================
        # 8b. CREA LINEA (POLYLINE) TRA SITO E PUNTO DI INTERESSE
        # ====================================================================
        
        polyline = folium.PolyLine(
            locations=[[site_lat, site_lon], [poi_lat, poi_lon]],
            color=line_color,
            weight=2.5,
            opacity=0.8,
            popup=f"""
            <b>{site_name}</b> → <b>{poi_name}</b><br>
            📏 Distanza: {distance:.2f} km<br>
            📍 Coordinate POI: ({poi_lat:.5f}, {poi_lon:.5f})
            """,
            tooltip=f"{poi_name}: {distance:.2f} km"
        )
        polyline.add_to(site_map)
        
        # ====================================================================
        # 8c. AGGIUNGI CERCHIO PER EVIDENZIARE IL PUNTO
        # ====================================================================
        
        folium.Circle(
            location=[poi_lat, poi_lon],
            radius=300,
            color=poi_color,
            fill=True,
            fill_color=poi_color,
            fill_opacity=0.4,
            popup=f"{poi_icon} {poi_name}<br>Distanza: {distance:.2f} km",
            tooltip=poi_name
        ).add_to(site_map)

# ============================================================================
# 9. AGGIUNGI LEGENDA COMPLETA
# ============================================================================

legend_html = '''
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: 340px; height: auto;
            border:2px solid grey; z-index:9999; font-size:12px;
            background-color:white; opacity:0.95; padding: 12px;
            border-radius: 8px;
            box-shadow: 3px 3px 5px rgba(0,0,0,0.2);
            font-family: Arial, sans-serif;">
<b>📍 LEGENDA COMPLETA</b><br>
<hr style="margin: 5px 0;">
🔴 <span style="color:red"><b>Sito di lancio</b></span> (con icona 🚀)<br>
🌊 <span style="color:blue"><b>Costa</b></span> - linea azzurra<br>
🏙️ <span style="color:green"><b>Città</b></span> - linea verde chiaro<br>
🚂 <span style="color:purple"><b>Ferrovia</b></span> - linea viola<br>
🛣️ <span style="color:orange"><b>Autostrada</b></span> - linea arancione<br>
⚓ <span style="color:darkblue"><b>Porto</b></span> - linea blu<br>
🏭 <span style="color:darkred"><b>Strutture</b></span> - linea rossa<br>
✈️ <span style="color:cadetblue"><b>Aeroporto</b></span> - linea cadetblue<br>
🖱️ <b>MousePosition</b> in alto a destra<br>
<br>
<b>📝 COME USARE:</b><br>
1. Zoom per avvicinarti ai siti<br>
2. Clicca su marker/polinea per info<br>
3. Muovi il mouse per coordinate<br>
4. Modifica le coordinate nei dizionari<br>
   <i>ccsfs_poi, ksc_poi, vafb_poi</i><br>
5. Riesegui la cella per aggiornare
</div>
'''

site_map.get_root().html.add_child(folium.Element(legend_html))

# ============================================================================
# 10. AGGIUNGI CONTROLLO STRATI (OPZIONALE)
# ============================================================================

folium.LayerControl().add_to(site_map)

# ============================================================================
# 11. MOSTRA MAPPA
# ============================================================================

print("\n" + "="*80)
print("🗺️ MAPPA CREATA CON SUCCESSO!")
print("="*80)
print("\n✅ Elementi aggiunti alla mappa:")
print("   • 🔴 Marker per i siti di lancio")
print("   • 📍 Marker con distanza per ogni punto di interesse")
print("   • 📏 Linee colorate tra siti e punti di interesse")
print("   • 🔵 Cerchi per evidenziare i punti")
print("   • 🖱️ MousePosition in alto a destra")
print("   • 📋 Legenda completa in basso a sinistra")
print("   • 🗂️ Controllo strati per attivare/disattivare")

# Mostra la mappa
display(site_map)

# ============================================================================
# 12. RIEPILOGO DISTANZE DETTAGLIATO
# ============================================================================

print("\n" + "="*80)
print("📊 RIEPILOGO DISTANZE DETTAGLIATO")
print("="*80)

for site_key, poi_dict in all_poi.items():
    site_lat = launch_sites[site_key]['lat']
    site_lon = launch_sites[site_key]['lon']
    site_name = launch_sites[site_key]['name']
    
    print(f"\n📍 {site_name}")
    print("-" * 50)
    
    distances_list = []
    for poi_name, poi_info in poi_dict.items():
        poi_lat = poi_info['lat']
        poi_lon = poi_info['lon']
        poi_icon = poi_info['icon']
        distance = calculate_distance(site_lat, site_lon, poi_lat, poi_lon)
        distances_list.append((poi_name, distance, poi_icon))
    
    # Ordina per distanza
    distances_list.sort(key=lambda x: x[1])
    
    for poi_name, distance, poi_icon in distances_list:
        print(f"   {poi_icon} {poi_name:<35} → {distance:>8.2f} km")

# ============================================================================
# 13. STATISTICHE GENERALI
# ============================================================================

print("\n" + "="*80)
print("📈 STATISTICHE GENERALI")
print("="*80)

all_distances = []
for site_key, poi_dict in all_poi.items():
    site_lat = launch_sites[site_key]['lat']
    site_lon = launch_sites[site_key]['lon']
    for poi_name, poi_info in poi_dict.items():
        poi_lat = poi_info['lat']
        poi_lon = poi_info['lon']
        distance = calculate_distance(site_lat, site_lon, poi_lat, poi_lon)
        all_distances.append(distance)

print(f"\n📊 Distanze minime: {min(all_distances):.2f} km")
print(f"📊 Distanze massime: {max(all_distances):.2f} km")
print(f"📊 Distanza media: {sum(all_distances)/len(all_distances):.2f} km")
print(f"📊 Totale punti di interesse: {len(all_distances)}")

print("\n" + "="*80)
print("✅ TASK COMPLETATO CON SUCCESSO!")
print("="*80)

print("\n💡 SUGGERIMENTI PER PERSONALIZZARE:")
print("   • Usa il MousePosition per trovare coordinate più precise")
print("   • Modifica i dizionari `ccsfs_poi`, `ksc_poi`, `vafb_poi`")
print("   • Aggiungi nuovi punti di interesse (stazioni, parchi, ecc.)")
print("   • Cambia i colori delle linee modificando 'line_color'")
print("   • Aggiungi icone personalizzate modificando 'icon'")
print("   • Riesegui la cella per aggiornare marker, distanze e linee")
```
